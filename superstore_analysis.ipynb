```python
import pandas as pd

# Load dataset
df = pd.read_csv('/content/SuperStoreOrders[1].csv', encoding='latin1')

# Fix column names
df.columns = df.columns.str.replace('Ã¯Â»Â¿', '')
df.columns = df.columns.str.strip().str.lower().str.replace(" ", "_")

# Cleaning
df.drop_duplicates(inplace=True)
df.fillna(0, inplace=True)

# â Handle mixed date formats (FINAL FIX)
df['order_date'] = pd.to_datetime(df['order_date'], format='mixed', dayfirst=True)
df['ship_date'] = pd.to_datetime(df['ship_date'], format='mixed', dayfirst=True)

# Check
print(df.dtypes)
df.head()
```
